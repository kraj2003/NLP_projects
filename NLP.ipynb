```python
corpus=" Hi , my name is khushi. I am learning NLP. NLP includes tokenization to convert it into smaller units.Tokenization is an important step in NLP."

```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
import nltk
nltk.download('all')
```

```python
# tokenization
# 
# import nltk as nltk
# nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
```

```python
documnets=sent_tokenize(corpus)
```

```python
type(documnets)
```

```python
for senence in documnets:
    print(senence)
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documnets:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
